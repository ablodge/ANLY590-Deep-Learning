```python
def relu(x):
    if x<0:
        return 0
    else:
        return x
    
def relu_deriv(x):
    return 0 if x<0 else 1

def leaky_relu(x, a=0.1):
    if x<0:
        return a*x
    else:
        return x
    
def leaky_relu_deriv(x, a=0.1):
    return a if x<0 else 1
```

```python
import matplotlib.pyplot as plt
xs = [x/100-0.5 for x in range(100)]
plt.plot(xs,[relu(x) for x in xs])
plt.show()
```

```python
xs = [x/100-0.5 for x in range(100)]
plt.plot(xs,[leaky_relu(x) for x in xs])
plt.show()
```
