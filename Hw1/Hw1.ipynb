## 1 Feedforward: Building a ReLu 2 Layer neural network


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def relu(x):
    return np.maximum(x,0)

def relu_derivative(x):
    return np.where(x>0, 1, 0)

def softmax(z):
    norm = np.sum(np.exp(z))
    return np.exp(z)/norm
```

### 1. Draw Network (see NN.jpg)

### 2. Equation

```python
%%latex
$y_{hat} = softmax\big(V*relu\big(W_2*relu\big(W_1*x+b_1\big)+b_2\big)+c\big)$
```

### 3. Python code `ff_nn_2_ReLu`

```python

def ff_nn_2_ReLu(x, parameters):
    W1, W2, V, b1, b2, c = parameters
    
    # Forward pass
    a1 = W1.dot(x) + b1                    # 3x1
    H1 = relu(a1)                          # 3x1
    a2 = W2.dot(H1) + b2                   # 2x1
    H2 = relu(a2)                          # 2x1
    a3 = V.dot(H2) + c                     # 3x1
    y_hat = softmax(a3)                    # 3x1
    return y_hat
    
```

### 4. Numeric example

```python
W1 = np.array([[1, 0], [-1, 0], [0, 0.5]]) # 3x2
W2 = np.array([[1, 0, 0], [-1, -1, 0]])    # 2x3
V = np.array([[1, 1], [0, 0], [-1, -1]])   # 3x2
b1 = np.array([[0,0,1]]).T                 # 3x1
b2 = np.array([[1,-1]]).T                  # 2x1
c = np.array([[1,0,0]]).T                  # 3x1

X = np.matrix([[1, 0, 0], [-1, -1, 1]])      # 2x3 (instance=column)
```

```python
params = (W1, W2, V, b1, b2, c)
x = X[:,0]
print('example 1')
print(ff_nn_2_ReLu(x, params))
x = X[:,1]
print('example 2')
print(ff_nn_2_ReLu(x, params))
x = X[:,2]
print('example 3')
print(ff_nn_2_ReLu(x, params))
```

Above, the reader can see probabilities associated with X=[[1,-1],[0,-1],[0,1]].

### 3. Gradient descent for Rosenbrock

```python
def rosenbrock(x,y):
    return (1-x)**2 + 100*(y-x**2)**2

def grad_rosenbrock(x,y):
    return np.array([-2+2*x-400*(y*x-x**3), 200*(y-x**2)])

def grad_descent(iterations=5, learning_rate=1e-2):
    starting_point = np.random.uniform(-0.1,0.1,size=2)
    trajectory = [starting_point]
    point = starting_point
    
    for i in range(iterations):
        grad = grad_rosenbrock(*point)
        point = point - learning_rate * grad
        trajectory.append(point)
    return (trajectory,[rosenbrock(x,y) for x,y in trajectory])
```

```python
iterations = 5
iters = [i for i in range(iterations+1)]

lr = 0.1
ys = grad_descent(learning_rate=lr)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr))
plt.show()
print(ys)

lr = 0.01
ys = grad_descent(learning_rate=lr)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr))
plt.show()
print(ys)

lr = 0.001
ys = grad_descent(learning_rate=lr)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr))
plt.show()
print(ys)
```

### 4. Gradient descent with momentum

```python
def grad_descent_momentum(iterations=5, learning_rate=1e-2, epsilon=0):
    starting_point = np.random.uniform(-0.1,0.1,size=2)
    trajectory = [starting_point]
    point = starting_point
    prev_grad = 0
    grad = 0
    
    for i in range(iterations):
        prev_grad = grad
        grad = grad_rosenbrock(*point)
        point = point - learning_rate * grad - epsilon*prev_grad
        trajectory.append(point)
    return (trajectory,[rosenbrock(x,y) for x,y in trajectory])
```

```python
iterations = 5
iters = [i for i in range(iterations+1)]

lr = 0.1
eps = 0.001
ys = grad_descent_momentum(learning_rate=lr,epsilon=eps)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr)+' epsilon: '+str(eps))
plt.show()


lr = 0.01
eps = 0.001
ys = grad_descent_momentum(learning_rate=lr,epsilon=eps)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr)+' epsilon: '+str(eps))
plt.show()

lr = 0.001
eps = 0.001
ys = grad_descent_momentum(learning_rate=lr,epsilon=eps)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr)+' epsilon: '+str(eps))
plt.show()

lr = 0.1
eps = 0.01
ys = grad_descent_momentum(learning_rate=lr,epsilon=eps)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr)+' epsilon: '+str(eps))
plt.show()


lr = 0.01
eps = 0.01
ys = grad_descent_momentum(learning_rate=lr,epsilon=eps)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr)+' epsilon: '+str(eps))
plt.show()

lr = 0.001
eps = 0.01
ys = grad_descent_momentum(learning_rate=lr,epsilon=eps)[1]
plt.plot(iters,ys)
plt.title('learning rate: '+str(lr)+' epsilon: '+str(eps))
plt.show()
print(ys)
```

Comparing the graphs for gradient descent with and without momentum, the reader can see that gradient descent on the Rosenbrock function only works with a small learning rate (around 0.01) and that momentum makes optimization worse. In some cases the optimized value explodes (as in graph 1 with learning rate: 0.1). 

## 3 Backpropagation

### 1. Gradients

```python
%%latex
\[y_{hat} = softmax\big(V*relu\big(W_2*relu\big(W_1*x+b_1\big)+b_2\big)+c\big)\]
\[\frac{\partial L}{\partial y_{hat}} = y_{hat}-y\]
\[~\]
\[\frac{\partial L}{\partial V} = (y_{hat}-y)\odot softmax'(a3)\cdot H2^T\]
\[~\]
\[\frac{\partial L}{\partial c} = (y_{hat}-y)\odot softmax'(a3)\cdot 1\]
\[~\]
\[\frac{\partial L}{\partial W2} = \big[V^T\big[(y_{hat}-y)\odot softmax'(a3)\big]\big]\odot relu'(a2)\cdot H1^T\]
\[~\]
\[\frac{\partial L}{\partial b2} = \big[V^T\big[(y_{hat}-y)\odot softmax'(a3)\big]\big]\odot relu'(a2)\cdot 1\]
\[~\]
\[\frac{\partial L}{\partial W1} = \big[W2^T\big[V^T\big[(y_{hat}-y)\odot softmax'(a3)\big]\big]\odot relu'(a2)\big]\odot relu'(a1)\cdot x^T\]
\[~\]
\[\frac{\partial L}{\partial b1} = \big[W2^T\big[V^T\big[(y_{hat}-y)\odot softmax'(a3)\big]\big]\odot relu'(a2)\big]\odot relu'(a1)\cdot 1\]
\[~\]


```

### 2. Python `grad_f`

```python
def grad_f(x, y, params):
    W1, W2, V, b1, b2, c = params
    
    a1 = W1.dot(x) + b1                    # d/dW1 = x, d/db1 = 1
    H1 = relu(a1)                          # d/da1 = (a1 > 0)
    a2 = W2.dot(H1) + b2                   # d/dW2 = H1, d/db2 = 1
    H2 = relu(a2)                          # d/da2 = (a2 > 0)
    a3 = V.dot(H2) + c                     # d/dV = H2, d/dc = 1
    yhat = softmax(a3)                    # d/da3 = ????
    
    d_loss = (yhat - y)                                                # 3x1
    d_V = d_loss.dot(H2.T)                                             # 3x2
    d_c = d_loss                                                       # 3x1
    d_W2 = (V.T.dot(d_loss) * (a2 > 0)).dot(H1.T)                      # 2x3
    d_b2 = (V.T.dot(d_loss) * (a2 > 0))                                # 2x1
    d_W1 = (W2.T.dot(V.T.dot(d_loss) * (a2 > 0)) * (a1 > 0)).dot(x.T)  # 3x2
    d_b1 = W2.T.dot(V.T.dot(d_loss) * (a2 > 0)) * (a1 > 0)             # 3x1
    
    return (d_W1, d_W2, d_V, d_b1, d_b2, d_c)
```

### 3. Bivariate Gaussians

```python
def plot_gmms(n = 999, plot=False):
    # based on multiclass-classification.ipynb (https://georgetown.instructure.com/files/1879711/download?download_frd=1)
    
    # Fixing seed for repeatability
    np.random.seed(333)
    
    # Parameters of a normal distribuion
    mean_1 = [0, 2] ; mean_2 = [2, -2] ; mean_3 = [-2, -2]
    mean = [mean_1, mean_2, mean_3] ; cov = [[1, 0], [0, 1]]  
    
    # Setting up the class probabilities
    n_samples = n
    pr_class_1 = pr_class_2 = pr_class_3 = 1/3.0
    n_class = (n_samples * np.array([pr_class_1,pr_class_2, pr_class_3])).astype(int)
  
    # Generate sample data
    for i in range(3):
        x1,x2 = np.random.multivariate_normal(mean[i], cov, n_class[i]).T
        if (i==0):
            xs = np.array([x1,x2])
            cl = np.array([n_class[i]*[i]])
        else: 
            xs_new = np.array([x1,x2])
            cl_new = np.array([n_class[i]*[i]])
            xs = np.concatenate((xs, xs_new), axis = 1)
            cl = np.concatenate((cl, cl_new), axis = 1)
    # Normalizing data (prevents overflow errors)     
    mu = xs.mean(axis = 1)
    std = xs.std(axis = 1)
    xs = (xs.T - mu) / std
    if (plot):
        plt.scatter(xs[:,0], xs[:,1], c=cl.flatten())
    else:
        onehot = {0:[1,0,0],1:[0,1,0],2:[0,0,1]}
        return xs, np.array([onehot[c] for c in cl.flatten()])
```

```python
plot_gmms(500,True)
```

### 4. Fit and Plot Loss

```python
def xentropy_loss(y, y_hat):
    # cross entropy
    tot = y * np.log(y_hat)
    return  -tot.sum()

def grad_descent(xs, ys, iterations=10, learning_rate=1e-2):
    s_W1  = np.random.uniform(-0.1,0.1,size=6).reshape((3,2))
    s_W2  = np.random.uniform(-0.1,0.1,size=6).reshape((2,3))
    s_V  = np.random.uniform(-0.1,0.1,size=6).reshape((3,2))
    s_b1  = np.random.uniform(-0.1,0.1,size=3).reshape((3,1))
    s_b2  = np.random.uniform(-0.1,0.1,size=2).reshape((2,1))
    s_c  = np.random.uniform(-0.1,0.1,size=3).reshape((3,1))
    start_params = (s_W1, s_W2, s_V, s_b1, s_b2, s_c)
    losses = []
    point = start_params
    
    for i in range(iterations):
        loss = 0
        for x,y in zip(xs,ys):
            x = x.reshape(2,1)
            y = y.reshape(3,1)
            W1, W2, V, b1, b2, c = point
            d_W1, d_W2, d_V, d_b1, d_b2, d_c = grad_f(x, y, point)
            W1 -= learning_rate * d_W1
            W2 -= learning_rate * d_W2
            V -= learning_rate * d_V
            b1 -= learning_rate * d_b1
            b2 -= learning_rate * d_b2
            c -= learning_rate * d_c
            point = W1, W2, V, b1, b2, c
            y_hat = ff_nn_2_ReLu(x, point)
            loss+=xentropy_loss(y, y_hat)
        losses.append(loss/len(ys))
    return (losses, point)
```

```python

iterations = 100
iters = [i for i in range(iterations)]

xs, ys = plot_gmms(500)
losses, params = grad_descent(xs,ys,iterations=iterations)
plt.plot(iters,losses)
plt.title('learning rate: 0.01')
plt.show()
```

### 5.  Fit and Plot Loss (Momentum)

```python
def grad_descent_momentum(xs, ys, iterations=10, learning_rate=1e-2, epsilon=0.5):
    s_W1  = np.random.uniform(-0.1,0.1,size=6).reshape((3,2))
    s_W2  = np.random.uniform(-0.1,0.1,size=6).reshape((2,3))
    s_V  = np.random.uniform(-0.1,0.1,size=6).reshape((3,2))
    s_b1  = np.random.uniform(-0.1,0.1,size=3).reshape((3,1))
    s_b2  = np.random.uniform(-0.1,0.1,size=2).reshape((2,1))
    s_c  = np.random.uniform(-0.1,0.1,size=3).reshape((3,1))
    start_params = (s_W1, s_W2, s_V, s_b1, s_b2, s_c)
    losses = []
    point = start_params
    prev_grad = grad = start_params # this is just small random numbers in the right shape
    
    for i in range(iterations):
        loss = 0
        for x,y in zip(xs,ys):
            prev_grad = grad
            p_W1, p_W2, p_V, p_b1, p_b2, p_c = prev_grad
            x = x.reshape(2,1)
            y = y.reshape(3,1)
            W1, W2, V, b1, b2, c = point
            grad = grad_f(x, y, point)
            d_W1, d_W2, d_V, d_b1, d_b2, d_c = grad
            W1 -= learning_rate * d_W1 + epsilon*p_W1
            W2 -= learning_rate * d_W2 + epsilon*p_W2
            V -= learning_rate * d_V + epsilon*p_V
            b1 -= learning_rate * d_b1 + epsilon*p_b1
            b2 -= learning_rate * d_b2 + epsilon*p_b2
            c -= learning_rate * d_c + epsilon*p_c
            point = W1, W2, V, b1, b2, c
            y_hat = ff_nn_2_ReLu(x, point)
            loss+=xentropy_loss(y, y_hat)
        losses.append(loss/len(ys))
    return (losses, point)
```

```python

iterations = 100
iters = [i for i in range(iterations)]

xs, ys = plot_gmms(500)
losses, params = grad_descent_momentum(xs,ys,iterations=iterations,epsilon=0.01)
plt.plot(iters,losses)
plt.title('learning rate: 0.01 epsilon: 0.01')
plt.show()
```

Gradient descent with momentum faster much better between 10 and 40 iterations (with a small epsilon value: 0.01). Outside of the 10-40 range, the two optimization methods are about the same.
